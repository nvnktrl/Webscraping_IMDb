```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
from time import sleep
from random import randint
```

```python
headers = {"Accept-Language": "en-US,en;q=0.5"}
```

```python
Movie_name = []
Year = []
Genre = []
Duration = []
Rating = []
Metascore = []
Votes = []
Gross = []
```

```python
pages = np.arange(1,1000,100)
```

```python
for page in pages:
    page = requests.get("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="+str(page)+"&ref_=adv_nxt")
    soup = BeautifulSoup(page.text, 'html.parser')
    Movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
    sleep(randint(2,8))
    for store in Movie_data:
        name = store.h3.a.text
        Movie_name.append(name)
    
        year = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
        Year.append(year)
    
        movie_genre = store.p.find('span', class_ = 'genre').text.replace('\n', '')
        Genre.append(movie_genre)
    
        runtime = store.p.find('span', class_ = 'runtime').text
        Duration.append(runtime)
    
        rating = store.find('div', class_ = 'inline-block ratings-imdb-rating').text.replace('\n', '')
        Rating.append(rating)
    
        metascore = store.find('span', class_ = 'metascore').text.replace('  ', '') if store.find('span', class_ = 'metascore') else ' '
        Metascore.append(metascore)
    
        values = store.find_all('span', attrs= {'name': 'nv'})
        vote = values[0].text
        Votes.append(vote)
    
        gross = values[1].text if len(values) > 1 else ''
        Gross.append(gross)
```

```python
Movies_df = pd.DataFrame({'Movie Name': Movie_name, 'Year': Year, 'Genre': Genre, 'Duration': Duration, 'Rating': Rating, 'Metascore': Metascore, 'Gross': Gross, 'Votes': Votes})    
```
